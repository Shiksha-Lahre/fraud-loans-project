```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("FraudLoanAnalysis") \
    .getOrCreate()
```

```python
fraud_df = spark.read.csv("fraud_transactions.csv", header=True, inferSchema=True)
fraud_df.toPandas().head()
```

```python
loan_df = spark.read.csv("loan_data.csv", header=True, inferSchema=True)
loan_df.toPandas().head()
```

```python
# Fraud class distribution
print("Fraud Class Distribution")
fraud_df.createOrReplaceTempView("fraud")
spark.sql("SELECT class, COUNT(*) as cnt FROM fraud GROUP BY class").show()

# Loan status distribution
print("Loan Class Distribution")
loan_df.createOrReplaceTempView("loans")
spark.sql("SELECT loan_status, COUNT(*) as cnt FROM loans GROUP BY loan_status").show()
```
